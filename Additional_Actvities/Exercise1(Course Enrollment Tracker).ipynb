```python
import sqlite3
import random

def run_exercise_1_with_warnings():
    conn = sqlite3.connect("enrollment.db")
    cursor = conn.cursor()

    # 1. Reset tables for clean data
    cursor.execute("DROP TABLE IF EXISTS enrollments")
    cursor.execute("DROP TABLE IF EXISTS students")
    cursor.execute("DROP TABLE IF EXISTS courses")

    cursor.execute("CREATE TABLE students (id INTEGER PRIMARY KEY, name TEXT)")
    cursor.execute("CREATE TABLE courses (id INTEGER PRIMARY KEY, name TEXT)")
    cursor.execute("""
        CREATE TABLE enrollments (
            student_id INTEGER,
            course_id INTEGER,
            PRIMARY KEY (student_id, course_id),
            FOREIGN KEY (student_id) REFERENCES students(id),
            FOREIGN KEY (course_id) REFERENCES courses(id)
        )
    """)

    # 2. Your specific Student and Course Data
    student_data = [
        (4168, "Jumar L. Sinues"), (4754, "John Richmond M. Cleofe"),
        (5870, "Kaye Celene G. Rivera"), (5552, "Gabriel T. Mañibo"),
        (4496, "Ayisha Claire R. Angeles"), (3560, "John Michael C. Alvarez"),
        (3570, "Trizz Ysabel V. Balingit"), (4172, "Jenny S. Guevarra"),
        (11919, "Ralph Lorenz N. Villafuerte"), (1234, "Jhon Mark Osorio")
    ]
    course_data = [(1, "BS ECE"), (2, "BS EE"), (3, "BS COE"), (4, "BS CE"), (5, "BS ME")]

    cursor.executemany("INSERT INTO students (id, name) VALUES (?, ?)", student_data)
    cursor.executemany("INSERT INTO courses (id, name) VALUES (?, ?)", course_data)

    # 3. THE WARNING LOGIC
    print("--- PROCESSING ENROLLMENTS ---")
    
    # First: Enroll everyone in BS ECE
    for s_id, s_name in student_data:
        try:
            cursor.execute("INSERT INTO enrollments (student_id, course_id) VALUES (?, ?)", (s_id, 1))
        except sqlite3.IntegrityError:
            print(f"Warning: {s_name} (ID: {s_id}) is already enrolled in BS ECE.")

    # Second: Fill up to 15 entries
    count = 10
    while count < 15:
        r_student = random.choice(student_data)
        r_sid = r_student[0]
        r_sname = r_student[1]
        r_cid = random.randint(2, 5)
        
        try:
            cursor.execute("INSERT INTO enrollments (student_id, course_id) VALUES (?, ?)", (r_sid, r_cid))
            count += 1
        except sqlite3.IntegrityError:
            # This triggers if the randomizer picks the same student/course combo again
            print(f"Warning: Duplicate enrollment attempt blocked for {r_sname} in Course ID {r_cid}")
            continue

    conn.commit()

     # 4. THE CONSOLIDATED OUTPUT (Grouped for your report)
    print("\n--- CONSOLIDATED STUDENT ENROLLMENT LIST ---")
    query = """
        SELECT students.id, students.name, GROUP_CONCAT(courses.name, ', ')
        FROM students
        JOIN enrollments ON students.id = enrollments.student_id
        JOIN courses ON enrollments.course_id = courses.id
        GROUP BY students.id
        ORDER BY students.name ASC
    """
    
    cursor.execute(query)
    for row in cursor.fetchall():
        print(f"ID: {row[0]} | Name: {row[1]:<30} | Courses: {row[2]}")

    conn.close()

run_exercise_1_with_warnings()
```
